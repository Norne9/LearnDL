### Convolution

```python
# load image
from PIL import Image

img = Image.open("images/steam-min.png")
img = img.convert("L")
img = np.array(img)
img = img / 255.0
```

```python
# create filter
x = np.linspace(-3, 3, 50)
fx = norm.pdf(x, loc=0, scale=1)
filt = np.outer(fx, fx)
plt.imshow(filt)
plt.show()
```

```python
# import convolve2d
from scipy.signal import convolve2d
```

```python
# apply filter
out = convolve2d(img, filt)
```

```python
# show result

plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(out)
plt.show()
```

### Exercise

```python
# load image
from PIL import Image

img = Image.open("images/steam-min.png")
img = img.convert("L")
img = np.array(img)
img = img / 255.0
```

```python
# filters
hx = np.array([[1,0,-1],
               [2,0,-2],
               [1,0,-1]])
hy = np.rot90(hx, k=-1)
hy
```

```python
# apply filters
from scipy.signal import convolve2d

gx = convolve2d(img, hx)
gy = convolve2d(img, hy)
```

```python
# final operation

g = np.sqrt(np.square(gx) + np.square(gy))
```

```python
# show result
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original")
plt.subplot(1,2,2)
plt.imshow(g)
plt.title("Edge detection")
plt.show()
```
