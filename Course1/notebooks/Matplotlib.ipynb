```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
plt.style.use("dark_background")
```

```python
# line plot
x = np.linspace(0, 20, 1000)
y = np.sin(x) + 0.2 * x

plt.plot(x, y)
plt.xlabel("input")
plt.ylabel("output")
plt.title("My plot")
plt.show()
```

```python
# scatter plot

x = np.random.randn(100, 2) # 2D array of rand numbers (rand - in range 0..1; randn - values around zero)
plt.scatter(x[:, 0], x[:, 1])
plt.show()

# same as
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y)
plt.show()
```

```python
# scatter plot with marks

x = np.random.randn(300, 2)
x[:100] += 2
x[100:200] += 4

marker = np.zeros(300)
marker[:100] = 1
marker[100:200] = 2

plt.scatter(x[:, 0], x[:, 1], c=marker)
plt.show()
```

```python
# histogram

x = np.random.randn(10000)
plt.hist(x)
plt.show()

# more bins
x = np.random.rand(10000)
plt.hist(x, bins=20)
plt.show()
```

```python
# load image

img = Image.open("images/steam-min.png") # loaded in unknown format
img = img.convert("RGB") # be sure its RGB
# other possible RGB, RGBA, HSV, L for grayscale, 1 for black/white
img = np.array(img)  # make array from image
img = img / 255.0  # convert from 0.255 to 0..1
img.shape  # array format - (y, x, channel)
```

```python
# show image

plt.imshow(img)
```

```python
# show green

# complicated green filter
green_mask = (img[:,:,1] > 0.1) & (img[:,:,1] > img[:,:,0]) & (img[:,:,1] > img[:,:,2])
# coordinates of true points
y, x = np.where(green_mask)

plt.imshow(img)
plt.scatter(x, y, c="g")
plt.show()
```

```python
# show in grayscale
# other cmap here - https://matplotlib.org/3.3.3/tutorials/colors/colormaps.html

plt.imshow(img.mean(axis=2), cmap="gray")
plt.show()
```
